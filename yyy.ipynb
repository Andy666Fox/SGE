```python
def conv(wav1: str, wav2: str) -> float:
    
    res = (np.sum(wav1) / np.sum(wav2)) * 100
    if res <= 0:
        return 0
    return res
```

```python
def compare(w1: str, w2: str):
    
    sr1, wav1 = wavfile.read(w1)
    sr2, wav2 = wavfile.read(w2)
    convergence = conv(wav1, wav2)
    
    fig, ax =   plt.subplots(figsize=(8,6))
    ax.plot(wav1, label=f'WAV1 srate: {sr1}')
    ax.plot(wav2, label=f'WAV2 srate:{sr2}')
    plt.title(f'Files convergence: {str(convergence)[:5]} % \n Percentage of losses: {str(100 - convergence)[:5]} %')
    plt.legend(loc='best')
    plt.show()  
```
